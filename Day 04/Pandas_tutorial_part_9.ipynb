```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first": ["Joey", "Chandler", "Ross", "Rachel", np.nan, None, "NA"],
    "last": ["Tribiyani", "Bing", "Geller", "Green", np.nan, np.nan, "Missing"],
    "email": ["howudoing@gmail.com", "waphaaa@gmail.com", "hiiii@gmail.com", None, np.nan, "moicagellerrrrr@gmail.com", "NA"],
    "age": [23, 24, 22, 22, None, None, "Missing"]
}
```

```python
df = pd.DataFrame(people)
df
```

### Use `dropna()` to drop rows/columns having missing values `NaN`

- `axis=0` or `axis="index"` by default
- `axis=1` or `axis="columns"` to drop missing value columns
- `how` parameter takes `any/all` to drop rows/columns contains any number of NaN values or **all** NaN values.
- `subset` takes a list of column on which it will check NaN to drop.
    - If only 1 column is provided, then `how` parameter will not be used.

```python
df.dropna()
```

```python
# Drop rows having "all" missing values
df.dropna(axis="index", how="all")
```

```python
# 'how' parameter is ignored because only one parameter
df.dropna(axis="index", subset=["email"])
```

```python
df.dropna(axis="index", how="all", subset=["last", "email"])
```

```python
df
```

```python
# Replacing "NA" or "Missing" with np.NaN
df.replace(["NA", "Missing"], np.NaN, inplace=True)
df
```

```python
# NaN is of float type, so integer columns will convert to float
type(np.NaN)
```

```python
df.dropna(axis="index", subset=["email"])
```

```python
df.dropna(axis="index", how="all", subset=["last", "email"])
```

### Use `isna()` to check where the `NaN` values are present

```python
df.isna()
```

### Use `fillena()` to fill missing values with provided value/series/dict

```python
df.fillna("MISSING")
```

```python
df.fillna({"age":0 , "first": "MISSING", "last": "MISSING", "email": "MISSING"})
```

```python
df.fillna({"age": 0}, inplace=True)
df
```

### Use `astype()` to typecast as series/columns of dataframe to different

```python
df["age"].astype(int)
```

```python
# converting types of age from float to int
df = df.astype({"age": int})
df
```
