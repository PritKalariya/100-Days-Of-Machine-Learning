### Calculate `median` over `mean` if there are outliers in the data

- mean is the average of numbers in the list.
- median is the middle number in the sorted list.
- **Outlier** is some data point that differs from other observations.
- Outlier is the specific data (can be high or low data) relative to the rest of the data. Like some persons have entered fake (high) salary, which can effect our rest of the data. So mean will include that salary while calculated. But, median doesn't include it.

### We can use `groupby()` to group results into a certain group

```python
df["Country"].value_counts()
```

```python
country_grp = df.groupby(["Country"])
```

```python
country_grp.get_group("India")
```

```python
# This is the same as above
df[df["Country"]=="India"]
```

```python
# How many people use which operating system
# based on each country or group by each country
country_grp["OpSys"].value_counts().head(20)
```

```python
# How many developers in India use which Operating System
df.loc[df["Country"]=="India"]["OpSys"].value_counts()
```

```python
# How many developers in India use which Operating System using groupby
country_grp["OpSys"].value_counts().loc["India"]
```

### Use `agg()` to do more than one operation

```python
country_grp["Salary"].agg(["median", "mean", "count", "max", "min"])
```

```python
# Getting Aggregate values of a specific country
country_grp["Salary"].agg(["mean", "median", "count", "min", "max"]).loc["India"]
```

```python
# Use 'str.contains' to find a specific value in a series
df[df["Country"]=="India"]["LanguageHaveWorkedWith"].str.contains("Python")
```

```python
# Counting total number of Indian Developers who know Python
df[df["Country"]=="India"]["LanguageHaveWorkedWith"].str.contains("Python").sum()
```

### Using `apply()` on Series Groupby object, to apply a specific function

```python
country_grp["LanguageHaveWorkedWith"].apply(lambda series: series.str.contains("Python").sum()).loc["India"]
```

### What percentage of people in each country knows Python

```python
no_of_response = df["Country"].value_counts()
no_of_response
```

```python
country_use_python = country_grp["LanguageHaveWorkedWith"].apply(lambda series: series.str.contains("Python").sum())
country_use_python
```

```python
solution_df = pd.concat([no_of_response, country_use_python], axis="columns")
solution_df
```

```python
solution_df.rename(columns={"Country": "TotalRespondents", "LanguageHaveWorkedWith": "KnowsPython"}, inplace=True)
solution_df
```

```python
# Calculating the % 
solution_df["Percentage"] = round(solution_df.KnowsPython / solution_df.TotalRespondents * 100, 2)
solution_df
```

```python
solution_df.sort_values(by=["Percentage", "TotalRespondents"], ascending=[False, False], inplace=True)
solution_df
```
