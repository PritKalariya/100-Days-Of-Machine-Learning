```python
import pandas as pd
from matplotlib import pyplot as plt, rcParams
```

```python
rcParams["figure.dpi"] = 100
plt.style.use("seaborn")
```

```python
%matplotlib inline
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

### Use `plt.scatter()` to create a scatter plot by passing `x` and `y` axis

```python
plt.scatter(x, y)
```

#### To increase size of markers, pass an integer value to `s`

```python
plt.scatter(x, y, s=100)
```

#### Pass a color `c` parameter to change the color of the markers.

#### We can change the marker by passing the marker style in `marker` parameter.

```python
plt.scatter(x, y, s=100, c="#0000FF", marker="^")
```

#### We  can change the edgecolour by passing a color to `edgecolor` parameter

- Also, change the width of edge by passing an integer to `linewidth` parameter.
- We can change the opacity of markers by passing a floating no to `aplha` parameter.

```python
plt.scatter(x, y, s=100, c="#00FF00",
           edgecolor="#0000FF", linewidth=1, alpha=0.5)
```

#### We can pass a list of different shades of color (in numbers) to represent each marker as each shade.

- Make sure the list of color shade numbers must be of smae size as `x` & `y`.


```python
x, y
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x, y, s=100, c=colors, 
            edgecolor="#0000FF", linewidth=1, alpha=0.5)
```

#### We can change the colors by different colormaps

- Use cmap paramter to set a colormap name in string
- Colormaps in Matplotlib are: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
def scatter_plot(*args, **kwargs):
    plt.scatter(x, y, edgecolor="#000000", linewidth=1, *args, **kwargs)
    plt.tight_layout()

scatter_plot(c=colors, cmap="cool", s=80)
```

#### Show the colormap `plt.colorbar()`

- Add label to colormap using `set_label()` method by passing a label

```python
plt.scatter(x, y, edgecolor="#000000", linewidth=1,
            c=colors, cmap="Greens", s=80)
plt.tight_layout()
color_bar = plt.colorbar()
color_bar.set_label("Satisfaction")
```

#### We can set different size for different marker by passing a list of all sizes in `s` parameter or set equal size by passing it an integer.

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, edgecolor="#000000", linewidth=1,
            c=colors, cmap="Greens", s=sizes)
plt.tight_layout()
color_bar = plt.colorbar()
color_bar.set_label("Satisfaction")
```

### Creating scatter plot of Top 200 Trending YT Videos

- Scatter plot is mostly used to see co-relations betwen the data
- On x-axis, we have `view counts`
- On y-axis, we have `likes`

```python
data = pd.read_csv("./data/yt-200trending.csv")
view_count = data["view_count"]
likes = data.likes
ratio = data.ratio
```

```python
plt.title("Top 200 Trending YT Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")
plt.scatter(view_count, likes, edgecolor="black", linewidth=1, alpha=0.5)
plt.tight_layout()
```

#### Set `xscale` & `yscale` to log if you have an outlier in the data.

- by setting xscale and yscale to log, we can see the data in logarithmic format.

```python
plt.title("Top 200 Trending YT Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")
plt.scatter(view_count, likes, edgecolor="black", linewidth=1, alpha=0.5)
plt.xscale("log")
plt.yscale("log")
plt.tight_layout()
```

```python
# Using ratio of likes/dislikes as colors
plt.title("Top 200 Trending YT Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")
plt.xscale("log")
plt.yscale("log")
plt.scatter(view_count, likes, c=ratio, cmap="autumn",
           edgecolor="black", linewidth=1, alpha=0.5)
cbar = plt.colorbar()
cbar.set_label("Like/Dislike Ratio")
plt.tight_layout()
```
