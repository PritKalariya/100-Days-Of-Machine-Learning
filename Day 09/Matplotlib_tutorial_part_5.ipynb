## Matplotlib Tutorial

### Part 5: Filling Area on Line Plots

**Source**: Corey Schafer [Matplotlib Playlist](https://www.youtube.com/playlist?list=PL-osiE80TeTvipOqomVEeZ1HRrcEvtZB_)

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
rcParams["figure.dpi"] = 100
plt.style.use("ggplot")
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("./data/developers_salary.csv")
data.head()
```

```python
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]
```

### Plotting a bassic line plot

```python
plt.plot(ages, dev_salaries, color="#FF0000", linestyle="--", label="All devs")
plt.plot(ages, py_salaries, color="#00FF00", marker=".", label="Python")
plt.legend()
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")
plt.tight_layout()
```

### We can use `plt.fill_between()` to fill area between two horizontal curves.

- Set `alpha` parameter to a float value, for the fill opacity.

```python
plt.plot(ages, dev_salaries, color="#0000FF", linestyle="--", label="All devs")
plt.plot(ages, py_salaries, color="#FF0000", marker=".", label="Python")
plt.legend()

plt.fill_between(ages, py_salaries, alpha=0.25)
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")
plt.tight_layout()
```

### Filling the gap between "All_Devs" and "Python" devs.

```python
plt.plot(ages, dev_salaries, color="#0000FF", linestyle="--", label="All devs")
plt.plot(ages, py_salaries, color="#FF0000", marker=".", label="Python")
plt.legend()

plt.fill_between(ages,dev_salaries, py_salaries, alpha=0.25)
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")
plt.tight_layout()
```

### Filling between the overall median salary and python dev salary.

```python
plt.plot(ages, dev_salaries, color="#0000FF", linestyle="--", label="All devs")
plt.plot(ages, py_salaries, color="#FF0000", marker=".", label="Python")
plt.legend()

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")
plt.tight_layout()
```

### We can fill the area with specific color, format, using `where` parameter.

- pass a condition in `where` parameter.
- Below, green area is showing python salaries greater than median salary.
- Red area is showing python salaries less than or equal to median salary

```python
plt.plot(ages, dev_salaries, color="#0000FF", linestyle="--", label="All devs")
plt.plot(ages, py_salaries, color="#FF0000", marker=".", label="Python")
plt.legend()

# python salaries greater than overall median salary
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25, color="green",
                 where=(py_salaries > overall_median), interpolate=True)

# python salaries less than or equal to overall median salary
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25, color="red",
                 where=(py_salaries <= overall_median), interpolate=True)

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")
plt.tight_layout()
```

### Filling gap between python devs salary & all devs salary

- We can also add `label` parameter to show in legend, whta these filled areas represent.

```python
plt.plot(ages, dev_salaries, color="#0000FF", label="All devs")
plt.plot(ages, py_salaries, color="#FF0000", label="Python")

# python salaries greater than overall median salary
plt.fill_between(ages, py_salaries, dev_salaries, alpha=0.25, color="green",
                 where=(py_salaries > dev_salaries), interpolate=True,
                 label="Above Avg.")

# python salaries less than or equal to overall median salary
plt.fill_between(ages, py_salaries, dev_salaries, alpha=0.25, color="red",
                 where=(py_salaries <= dev_salaries), interpolate=True,
                 label="Below Avg.")

plt.legend()
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")
plt.tight_layout()
```
