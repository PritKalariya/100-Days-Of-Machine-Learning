### Use `plt.pie()` to plot a pie chart

- Pass an array of numbers/slices as 1st parameter or assign to `x`.
- Set an array for `labels` paramete and pass labels according to slices.
- Each portin is a slice in pie chart.

```python
slices = [60, 40]
labels = ["Sixty", "Forty"]
plt.title("My Awesome Pie Chart")
plt.pie(slices, labels=labels)
```

### Pass a dictionary of properties for the pie chart to wedgeprops

- List of attributes is [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Wedge.html)

```python
slices = [60, 40]
labels = ["Sixty", "Forty"]
plt.title("My Awesome Pie Chart")
plt.pie(slices, labels=labels, wedgeprops={"edgecolor": "black"})
```

### Pass a list of colors in `colors` parameter

- colors can be color names or hex values

```python
# numbers can be of any range, matplotlib will automatically sets according to the ratio
slices = [120, 80, 30, 20]
labels = ["Sixty", "Forty", "Extra1", "Extra2"]
colors = ["#00CCFF", '#FFCCFF', "#FFFFCC", "#00FFCC"]
plt.title("My Awesome Pie Chart")
plt.pie(slices, labels=labels, colors=colors, wedgeprops={"edgecolor":"black"})
plt.show()
```

### Plotting Pie Chart on language popularity data

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/language_popularity.csv")
df.sort_values(by="Popularity", ascending=False, inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
df.head()
```

```python
popularity = df["Popularity"][:5]
languages = df["Language"][:5]
plt.title("Top 5 popular languages in 2021")
plt.pie(popularity, labels=languages, wedgeprops={"edgecolor":"black"})
plt.show()
```

### Use `explode` parameter to offset a specific slice of a pie chart

- pass an array of numbers from 0 to 1.
- 0.5 will offset the slice half of radius away.

```python
explode = [0, 0, 0.1, 0, 0.5]
plt.pie(popularity, labels=languages, explode=explode,
        wedgeprops={"edgecolor":"black"})
plt.show()
```

### Use `shadow` parameter to give 3d look to pie chart

- set `shadow=True`

```python
plt.title("Popular Languages in 2021")
plt.pie(popularity, labels=languages, explode=explode,
        wedgeprops={"edgecolor":"black"}, shadow=True)
plt.show()
```

### Use `startangle` parameter to rotate pie chart with some degree

- pass an integer to it as degree

```python
plt.title("Popular Languages in 2021")
plt.pie(popularity, labels=languages, explode=explode,
        wedgeprops={"edgecolor":"black"}, shadow=True,
        startangle=-90)
plt.show()
```

### Use `autopct` parameter and set a format string to see percentage in the pie chart

- `%.2f` for formatted percentage
- `%%` at end to show the percentage sign

```python
plt.title("Popular Languages in 2021")
plt.pie(popularity, labels=languages, explode=explode,
        wedgeprops={"edgecolor":"black"}, shadow=True,
        startangle=-90, autopct="%.2f%%")
plt.show()
```
